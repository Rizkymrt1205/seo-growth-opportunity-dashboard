## Identifikasi Kolom yang Akan Dipakai

```python
seo_df = df[
    [
        "Date",
        "Primary Keywords",
        "Location",
        "Device Type",
        "Organic_Traffic",
        "Keywords_Ranking",
        "Clicks",
        "Impressions",
        "CTR (%)",
        "Average_Position",
        "Conversion_Rate (%)",
        "Top_Landing_Pages",
        "Keyword_Difficulty",
        "Organic_Revenue ($)",
        "Technical_SEO_Errors"
    ]
]

seo_df.head()
```

## Buat Opportunity Score

```python
seo_df["Opportunity_Score"] = (
    seo_df["Impressions"] *
    (1 / seo_df["Average_Position"])
)
```

```python
seo_df[
    [
        "Primary Keywords",
        "Impressions",
        "Average_Position",
        "Opportunity_Score"
    ]
].head()
```

```python
## Buat Action Category
```

```python
def classify_position(position):

    if position <= 3:
        return "Maintain"

    elif position <= 10:
        return "Quick Win"

    elif position <= 20:
        return "Growth Opportunity"

    else:
        return "Low Priority"
```

```python
seo_df["Action_Category"] = seo_df[
    "Average_Position"
].apply(classify_position)
```

```python
seo_df[
    [
        "Primary Keywords",
        "Average_Position",
        "Action_Category"
    ]
].head()
```

```python
seo_df["Action_Category"].value_counts()
```

```python
seo_df.to_csv(
    "../data_clean/seo_dashboard_final.csv",
    index=False
)

print("Export berhasil")
```

## pahami cerita datanya dulu

```python
seo_df.groupby("Action_Category").agg({
    "Clicks": "sum",
    "Impressions": "sum",
    "Organic_Traffic": "sum",
    "Organic_Revenue ($)": "sum"
}).round(2)
```

```python
top_opportunity = (
    seo_df
    .sort_values(by="Opportunity_Score", ascending=False)
)

top_opportunity[
    [
        "Primary Keywords",
        "Impressions",
        "Average_Position",
        "Opportunity_Score",
        "Action_Category"
    ]
].head(20)
```
